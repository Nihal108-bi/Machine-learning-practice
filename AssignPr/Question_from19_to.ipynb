#Question 20. What do you mean by Measure of Central Tendency and Measures of Dispersion .How it can be calculated.  
#ANS:

<<<Measures of Central Tendency>>>
Measures of central tendency are statistical metrics used to identify a central point or typical value within a dataset. They help summarize a large dataset by providing a single value that represents the center of the data. The most common measures of central tendency are:

1. Mean: The average of all data points.
Calculation: Sum of all data points divided by the number of data points.

2. Median: The middle value when data points are arranged in ascending or descending order.
Calculation:
If the number of data points ((n)) is odd, the median is the middle value.
If (n) is even, the median is the average of the two middle values.

3. Mode: The most frequently occurring value in the dataset.
Calculation: Identify the value that appears most frequently.

<<<Measures of Dispersion>>>
Measures of dispersion describe the spread or variability of a dataset. They indicate how much the data points differ from the central tendency.

1. Range: The difference between the maximum and minimum values.
2. Variance: The average of the squared differences from the mean.
3. Standard Deviation: The square root of the variance, providing a measure of dispersion in the same units as the data.
4. Interquartile Range (IQR): The range between the first quartile (Q1) and the third quartile (Q3).



```python
#Calculation methods in pyhton

import numpy as np
from scipy import stats

#samplle data
data=[2,3,4,4,5,11,13,17]

#mean
print(f"mean is :{np.mean(data)}")

#median
print(f"median is :{np.median(data)}")

#mean
print(f"mode is :{stats.mode(data)}")

#Range
print(f"range is :{np.ptp(data)}")

#Variance
print(f"variance is :{np.var(data)}")

#Standard deviation
print(f"mean is :{np.std(data)}")

#Interquartile Range(IQR)
Q1=np.percentile(data,25)
Q3=np.percentile(data,75)
print(f"IQR is: {Q3-Q1}")
```

# Question 21. What do you mean by skewness.Explain its types.Use graph to show. 
# Answer:-

<<Skewness>>
Definition: Skewness is a measure of the asymmetry of the probability distribution of a real-valued random variable about its mean. It indicates whether the data points are skewed to the left (negative skew) or to the right (positive skew).

Types of Skewness:

1. Positive Skew (Right Skew):
The right tail (higher values) is longer or fatter than the left tail.
Most of the data points are concentrated on the left side.
Mean > Median > Mode.

2. Negative Skew (Left Skew):
The left tail (lower values) is longer or fatter than the right tail.
Most of the data points are concentrated on the right side.
Mean < Median < Mode.

3. Zero Skew (Symmetrical Distribution):
The tails on both sides of the mean are balanced.
The distribution is symmetrical.
Mean = Median = Mode.

# Question 22. Explain PROBABILITY MASS FUNCTION (PMF) and PROBABILITY DENSITY FUNCTION (PDF). and what is the difference between them? 
# Answer:

<<<Probability Mass Function (PMF)>>>
Definition: 
The Probability Mass Function (PMF) is a function that gives the probability that a discrete random variable is exactly equal to some value.

Usage: 
PMF is used for discrete random variables, which can take on a finite or countably infinite number of values.

Properties:
>The sum of all probabilities in the PMF is 1.
>Each probability value is between 0 and 1.

<<<Probability Density Function (PDF)>>>
Definition:
The Probability Density Function (PDF) is a function that describes the likelihood of a continuous random variable to take on a particular value.

Usage: 
PDF is used for continuous random variables, which can take on an infinite number of values within a given range.

Properties:
>The area under the PDF curve over the entire range of possible values is 1.
>The value of the PDF at any given point is not a probability but a density.
>To find the probability that the variable falls within a certain range, you integrate the PDF over that range.

<<<Differences Between PMF and PDF>>>
1. Type of Variable:
>PMF is used for discrete random variables.
>PDF is used for continuous random variables.

2. Probability Calculation:
>PMF gives the probability that a random variable is exactly equal to a specific value.
>PDF gives the density of the probability at a specific value, and probabilities are found by integrating the PDF over an interval.

3. Sum vs. Integral:
>The sum of all probabilities in a PMF is 1.
>The integral of the PDF over the entire range is 1.

4. Value Interpretation:
>PMF values are actual probabilities.
>PDF values are densities and not probabilities themselves.

# Question 28. What is  Normal Distribution? What are the four Assumptions of Normal Distribution? Explain in detail. 
# Answer:

**Normal Distribution** is a fundamental concept in statistics that describes how data values are distributed. It is characterized by its symmetric, bell-shaped curve, which is defined by its mean (average) and standard deviation (spread). The normal distribution is also known as the Gaussian distribution, after the mathematician Carl Friedrich Gauss.

### Characteristics of Normal Distribution

1. **Symmetry**: The distribution is perfectly symmetric around its mean. This means that the left and right halves of the distribution are mirror images of each other.
   
2. **Mean, Median, and Mode**: In a normal distribution, the mean, median, and mode are all equal and located at the center of the distribution.
   
3. **Bell-Shaped Curve**: The graph of a normal distribution forms a bell-shaped curve. It is highest at the mean and decreases as you move away from the center.

4. **68-95-99.7 Rule**: Approximately 68% of the data falls within one standard deviation of the mean, 95% within two standard deviations, and 99.7% within three standard deviations.

### Four Assumptions of Normal Distribution

1. **Symmetry**:
   - **Assumption**: The distribution of the data should be symmetric around the mean. This means that the distribution has the same shape on both sides of the mean.
   - **Explanation**: In a perfectly normal distribution, if you were to fold the curve along the mean, both halves would overlap exactly. This symmetry ensures that the mean, median, and mode are the same.
   - **Example**: The heights of adult men in a population often follow a normal distribution, where most men are of average height, with fewer men being extremely short or tall.

2. **Mean and Variance**:
   - **Assumption**: The distribution is described by its mean () and variance (). The mean determines the location of the center of the distribution, and the variance determines the spread or width of the distribution.
   - **Explanation**: In a normal distribution, the spread of the data is determined by the standard deviation (), which is the square root of the variance. The mean () shifts the curve along the horizontal axis.
   - **Example**: If the mean score on an exam is 75 and the variance is 25, the distribution will be centered around 75 with a standard deviation of 5.

3. **Independence**:
   - **Assumption**: The observations should be independent of each other. This means that the value of one observation does not influence the value of another observation.
   - **Explanation**: Independence ensures that the probability of observing a value in one part of the distribution does not affect the probability of observing a value in another part.
   - **Example**: When rolling a fair die multiple times, each roll is independent of the others, and the outcomes should follow a normal distribution when aggregated.

4. **Infinite Number of Observations**:
   - **Assumption**: For the theoretical properties of the normal distribution to hold, the sample size should be large enough, ideally approaching infinity. In practice, a sample size of 30 or more is often considered sufficient.
   - **Explanation**: The normal distribution is a continuous distribution, and in real-world applications, we use the normal distribution to approximate the behavior of sample means or sums from large samples.
   - **Example**: When measuring the average height of a large number of individuals, the sample mean will approximate a normal distribution due to the Central Limit Theorem, even if the individual heights are not normally distributed.



# Question 29.Write all the characteristics or Properties of the Normal Distribution Curve. 
# Answer:
### Characteristics or Properties of the Normal Distribution Curve

The normal distribution, also known as the Gaussian distribution, has several key characteristics that define its shape and behavior. Here are the main properties:

1. **Symmetry**:
   - **Description**: The normal distribution curve is perfectly symmetric around its mean. The left and right sides of the curve are mirror images of each other.
   - **Implication**: The mean, median, and mode of the distribution are all equal and located at the center of the distribution.

2. **Bell-Shaped Curve**:
   - **Description**: The graph of a normal distribution forms a bell-shaped curve, which is highest at the mean and tapers off as you move away from the mean.
   - **Implication**: The curve decreases exponentially as you move further from the mean.

3. **Mean, Median, and Mode**:
   - **Description**: In a normal distribution, the mean, median, and mode are all equal and located at the peak of the curve.
   - **Implication**: This equality reflects the symmetry of the distribution.

4. **Asymptotic Nature**:
   - **Description**: The tails of the normal distribution curve approach, but never actually touch, the horizontal axis. They extend infinitely in both directions.
   - **Implication**: This means that extreme values are possible, though they become increasingly rare as you move further from the mean.

5. **68-95-99.7 Rule (Empirical Rule)**:
   - **Description**: Approximately 68% of the data falls within one standard deviation of the mean, 95% within two standard deviations, and 99.7% within three standard deviations.
   - **Implication**: This rule provides a quick way to understand the spread of data in a normal distribution.

6. **Defined by Mean and Standard Deviation**:
   - **Description**: The shape of the normal distribution is determined by its mean () and standard deviation (). The mean determines the location of the center of the distribution, and the standard deviation determines the width of the bell curve.
   - **Implication**: Different means shift the curve along the horizontal axis, while different standard deviations affect the spread of the curve.

7. **Total Area Under the Curve**:
   - **Description**: The total area under the normal distribution curve is equal to 1. This reflects the fact that the probability of all possible outcomes sums to 100%.
   - **Implication**: This property is used in probability calculations and statistical inference.

8. **Central Limit Theorem**:
   - **Description**: The Central Limit Theorem states that the distribution of sample means approaches a normal distribution as the sample size becomes large, regardless of the original distribution of the data.
   - **Implication**: This property underpins many statistical methods and hypothesis tests that rely on normality.

9. **Standard Normal Distribution**:
    - **Description**: A special case of the normal distribution is the standard normal distribution, which has a mean of 0 and a standard deviation of 1.
    - **Implication**: It is used to simplify calculations and is referenced using z-scores, which standardize values from any normal distribution.

# Question34. What is the statistical hypothesis? Explain the errors in hypothesis testing.b)Explain the  Sample. What are Large Samples & Small Samples?
# Answer:
### Statistical Hypothesis

A **statistical hypothesis** is a specific statement or assumption about a population parameter that can be tested using statistical methods. It represents a claim about a characteristic of a population, such as a mean or proportion, and provides a basis for testing whether the observed data provides sufficient evidence to support or refute this claim.

#### Types of Hypotheses

1. **Null Hypothesis (\(H_0\))**:
   - **Description**: A statement of no effect or no difference. It represents the default assumption that there is no relationship or effect.
   - **Example**: "There is no difference in the average test scores between two teaching methods."

2. **Alternative Hypothesis (\(H_1\) or \(H_a\))**:
   - **Description**: A statement that contradicts the null hypothesis. It represents the presence of an effect or a difference.
   - **Example**: "There is a difference in the average test scores between two teaching methods."

### Errors in Hypothesis Testing

In hypothesis testing, there are two main types of errors that can occur:

1. **Type I Error (False Positive)**:
   - **Description**: Occurs when the null hypothesis (\(H_0\)) is rejected when it is actually true. In other words, it is the incorrect conclusion that there is an effect or difference when there is none.
   - **Probability**: Represented by the significance level (\(\alpha\)), often set at 0.05.
   - **Example**: Concluding that a new drug is effective when it is actually not effective.

2. **Type II Error (False Negative)**:
   - **Description**: Occurs when the null hypothesis (\(H_0\)) is not rejected when it is actually false. This means failing to detect an effect or difference that does exist.
   - **Probability**: Represented by \(\beta\), and \(1 - \beta\) is known as the power of the test.
   - **Example**: Concluding that a new drug is not effective when it actually is effective.

### Sample

A **sample** is a subset of individuals or observations selected from a larger population. The purpose of sampling is to make inferences about the population based on the data from the sample.

#### Large Samples vs. Small Samples

1. **Large Samples**:
   - **Description**: Samples that contain a relatively large number of observations. The exact size that constitutes a large sample can vary, but generally, a sample size greater than 30 is considered large.
   - **Advantages**:
     - Provides more accurate and reliable estimates of population parameters.
     - Reduces the margin of error and increases the statistical power of hypothesis tests.
     - Often results in a sample distribution that approximates the normal distribution due to the Central Limit Theorem.
   - **Example**: A survey of 1,000 people from a population of 100,000.

2. **Small Samples**:
   - **Description**: Samples with a relatively small number of observations. Typically, a sample size less than 30 is considered small.
   - **Disadvantages**:
     - May provide less reliable estimates of population parameters.
     - Higher margin of error and less statistical power in hypothesis tests.
     - Sample distributions may not approximate normality, making some statistical tests less valid.
   - **Example**: A clinical trial with 20 participants.
