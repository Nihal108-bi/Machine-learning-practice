```python
#Multivarite analysis

cols=['AGE','SMOKING','FATIGUE','ALLERGY','ALCOHOL_CONSUMING','CHEST_PAIN','LUNG_CANCER']

g=sns.PairGrid(data[cols],hue='LUNG_CANCER',palette='coolwarm')
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot, cmap = 'Blues_d')
g.map_diag(sns.histplot)
g.add_legend()
plt.title("PairGrid for selected columns")
plt.show()
```
